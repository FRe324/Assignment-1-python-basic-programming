Write a Python program to swap two variables?

```python
#to swap two no.
val1=int(input("enter no. 1 "))
val2=int(input("enter no. 2 "))
t=val1
val1=val2
val2=t
print("the swapped value of no 1",val1)
print("the swapped value of no 2",val2)
```
